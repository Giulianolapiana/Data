Ejercicio 1: Slicing y Sub-matrices

```python
m = np.arange(1,26)
m = m.reshape(5,5)
print("Matriz original:\n", m)
m = m[1:4, 1:4]
print("Matriz recortada a 4x4:\n", m)
```

Ejercicio 2: Estadísticas con Ejes (Axis)

```python
ventas = np.array([
    [1500, 2000, 1800, 2200],
    [1100, 1300, 1200, 1500],
    [3000, 3100, 2900, 3500]
])
total = np.sum(ventas)
print("Venta total:", total)
promedio_semanal = np.mean(ventas, axis=0)
print("Promedio por semana:", promedio_semanal)
max_sucursal = np.max(ventas, axis=1)
print("Máximo por sucursal:", max_sucursal)
```
