Ejercicio 1

```python
import pandas as pd
import numpy as np
# Generamos 50 registros simulando el reporte del servidor
np.random.seed(77)
total_registros = 50
cursos_disponibles = ['Python Básico', 'Bases de Datos', 'Data Science', 'Desarrollo Web']
datos_plataforma = {
    'ID_Estudiante': ['EST-' + str(i) for i in np.random.randint(1000, 1050, total_registros)],
    'Curso': np.random.choice(cursos_disponibles, total_registros),
    'Minutos_Vistos': np.random.randint(10, 600, total_registros),
    'Nota_Examen': np.random.uniform(2.0, 10.0, total_registros).round(1)
}
df_generador = pd.DataFrame(datos_plataforma)
# Inyectamos el caos (Errores de sistema)
# 1. Duplicamos las primeras 5 filas
df_generador = pd.concat([df_generador, df_generador.iloc[0:5]], ignore_index=True)
# 2. Borramos 3 IDs de estudiante (Nulos críticos)
# 2. Borramos 3 IDs de estudiante (Nulos críticos)
df_generador.loc[[12, 25, 41], 'ID_Estudiante'] = np.nan
# 3. Borramos 8 notas de examen (Alumnos ausentes)
df_generador.loc[[3, 8, 15, 22, 29, 35, 44, 48], 'Nota_Examen'] = np.nan
# 4. Borramos 5 nombres de cursos (Error de registro)
df_generador.loc[[5, 18, 27, 33, 49], 'Curso'] = np.nan
# Desordenamos y exportamos
df_generador = df_generador.sample(frac=1).reset_index(drop=True)
df_generador.to_csv('reporte_elearning.csv', index=False)
print('Archivo reporte_elearning.csv generado. ¡A limpiar!')
```
