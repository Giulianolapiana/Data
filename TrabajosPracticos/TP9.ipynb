Ejercicio 1

```python
import pandas as pd
import numpy as np

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
np.random.seed(42)

# Generamos datos de los últimos 12 meses
datos_it = {
    'Mes': meses,
    'Sueldo_Promedio_USD': [1100, 1150, 1200, 1180, 1250, 1300, 1350, 1400, 1380, 1450, 1500, 1550],
    'Puestos_Abiertos': np.random.randint(50, 200, 12)
}

df_it = pd.DataFrame(datos_it)
df_it.to_csv('mercado_it_2025.csv', index=False)

print('Archivo mercado_it_2025.csv generado con éxito.')
```

Ejercicio 2

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("mercado_it_2025.csv")
print(df.head())
```

```python
plt.figure(figsize=(10, 5))

plt.plot(
    df["Mes"],
    df["Sueldo_Promedio_USD"],
    color="green",
    marker="o"
)

plt.title("Evolución del sueldo promedio IT en 2025")
plt.xlabel("Mes")
plt.ylabel("Sueldo Promedio (USD)")
plt.grid(True)

plt.show()
```

Ejercicio 3

```python
plt.figure(figsize=(10, 5))

plt.bar(
    df["Mes"],
    df["Puestos_Abiertos"],
    color="orange"
)

plt.title("Evolución de puestos abiertos en IT en 2025")
plt.xlabel("Mes")
plt.ylabel("Puestos Abiertos")
plt.grid(True)

plt.show()
```

```python
puestos_max = df["Puestos_Abiertos"].max()
for i in range(len(df)):
    if df["Puestos_Abiertos"][i] == puestos_max:
        mes_max = df["Mes"][i]
        break

print(f"El mes con mayor demanda de puestos fue {mes_max} con {puestos_max} puestos abiertos.")
```

Ejercicio 4

```python
plt.figure(figsize=(12, 6))
plt.style.use('ggplot')
plt.plot(
    df["Mes"],
    df["Sueldo_Promedio_USD"],
    color="blue"
)
valor_promedio = df["Sueldo_Promedio_USD"].mean()
plt.axhline(y=valor_promedio, color='red', linestyle='--', label=f'Promedio: {valor_promedio:.2f} USD')
plt.title("Evolución del sueldo promedio IT en 2025")
plt.xlabel("Mes")
plt.ylabel("Sueldo Promedio (USD)")
plt.grid(True)
plt.show()
```
