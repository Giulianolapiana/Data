```python
import pandas as pd
import numpy as np

# Generación de datos simulados
np.random.seed(123)

legajos_generados = ['TUP-' + str(i) for i in range(1001, 1021)]

carreras_disponibles = ['Programacion', 'Redes', 'Sistemas']

datos_academicos = {
    'Legajo': legajos_generados,
    'Carrera': np.random.choice(carreras_disponibles, 20),
    'Matematica': np.random.randint(2, 11, 20),
    'Bases_Datos': np.random.randint(2, 11, 20),
    'Algoritmos': np.random.randint(2, 11, 20)
}

df_generador = pd.DataFrame(datos_academicos)

df_generador.to_csv('notas_semestre.csv', index=False)

print('Archivo notas_semestre.csv generado con éxito.')
```

```python
import pandas as pd

df_notas = pd.read_csv("notas_semestre.csv")
```
